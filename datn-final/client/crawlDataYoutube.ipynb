```python
from googleapiclient.discovery import build

# Replace with your own API key
api_key = 'AIzaSyBIvXUMdQ0UlXnsRtA2FS4s2IRYqykoXLI'

# Create a YouTube API client
youtube = build('youtube', 'v3', developerKey=api_key)

# Function to search for the latest videos by keyword
def search_latest_videos_by_keyword(keyword, max_results=10):
    request = youtube.search().list(
        q=keyword,
        part='snippet',
        type='video',
#         order='viewCount',  # Order by date to get the latest videos
        maxResults=max_results
    )
    response = request.execute()

    video_urls = []
    for item in response['items']:
        video_id = item['id']['videoId']
        video_url = f'https://www.youtube.com/watch?v={video_id}'
        video_urls.append(video_url)
        title = item['snippet']['title']
        description = item['snippet']['description']
        publish_time = item['snippet']['publishedAt']
        print(f'Title: {title}')
        print(f'Description: {description}')
        print(f'URL: {video_url}')
        print(f'Published at: {publish_time}\n')

    return video_urls

# Example usage
keyword = 'quản lý tài chính'
max_results = 10  # Adjust the number of results as needed

print('Search Results:')
search_latest_videos_by_keyword(keyword, max_results)
```
